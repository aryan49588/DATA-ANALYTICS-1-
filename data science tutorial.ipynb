```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
covid = pd.read_csv('covid_19_india+%281%29.csv',parse_dates=['Date'],dayfirst=True)
covid.head()
```

```python
covid = covid[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]

covid.columns = ['date','state','cured','deaths','confirmed']
covid
```

```python
covid.head(30)
```

```python
covid.tail()
```

```python
today=covid[covid['date']=='2021-04-22	']
today
```

```python
max_cases=today.sort_values(by="confirmed",ascending=False)
max_cases
```

```python
top_states_confirmed = max_cases[0:10]
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_deaths=today.sort_values(by="deaths",ascending=False)
max_deaths
```

```python
top_state_death=max_deaths[0:10]
g=sns.barplot(x="state",y="deaths",data=top_state_death,hue='state')
g
```
